```python
def Register():
    print("------------------------------------------------------------")
    print("Welcome to Kamal Bank! Register your account here")
    global username
    global password
    username = input("Enter your username: ")
    password = input("Enter your password: ")
    print("Thanks for making account here! Login with your details now")
    print("------------------------------------------------------------")


def auth():
    print("------------------------------------------------------------")
    us = input("Username: ")
    pw = input("password: ")
    if us == username and pw == password:
        print("------------------------------------------------------------")
        print("Login successful")
        print("------------------------------------------------------------")
        return True
    else:
        print("------------------------------------------------------------")
        print("Invalid username or password")
        print("------------------------------------------------------------")
        return False


def welcome(choice):
    global balance
    if choice == 1:
        return f"Current balance is {balance}"
    elif choice == 2:
        deposit = int(input("How much amount you want to deposit: "))
        balance += deposit
        return f"deposited {deposit}"
    elif choice == 3:
        withdraw = int(input("How much balance you want to Withdraw: "))
        if withdraw >= balance:
            return("Low balance")
        else:
            balance -= withdraw
            return f"Withdrawn {withdraw}"
    else:
        return "Bye Bye!"


flag = True
username = ''
password = ''
balance = 0
Register()
auth = auth()
if auth:
    print("------------------------------------------------------------")
    print("Welcome to Kamal Bank")
    print("Enter your choice: \n1. Account Balance\n2. Deposit\n3. Withdraw\n4. Exit")
    while flag:
        print("------------------------------------------------------------")
        choise = int(input('Enter your choise: '))
        print("------------------------------------------------------------")
        print(welcome(choise))
        if choise == 4:
            flag = False
            break
else:
    print("------------------------------------------------------------")
```
